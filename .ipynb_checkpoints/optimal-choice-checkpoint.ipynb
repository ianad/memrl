```python
import os
import pandas as pd
import numpy as np
```

```python
sourcedir = os.path.join('..','sourcedata')
outputdir = os.path.join('..','derivatives')
```

```python
fpath = os.path.join(sourcedir,'memrl_data.xlsx')
data = pd.read_excel(fpath)
```

```python
data.head()
```

```python
df={}
for sub in data['SubjNum'].unique():
    for cue in data['CueChosen'].unique():
        key = (str(sub) +'-'+ str(cue))
        df[key] = data[data['SubjNum'] == sub][data['CueChosen'] == cue]
```

```python
df['101-kanji7.jpg']
```

```python
key = '101-kanji2.jpg'
columns = ['SubjNum','CueChosen','Outcome',
           'PreviousPositiveOutcomes','PreviousTotalOutcomes',
           'PositiveOutcomeFrequency','ChoiceOptimal']
df[key]['PreviousPositiveOutcomes'] = df[key]['Outcome'].cumsum().shift(1)
df[key]['PreviousTotalOutcomes'] = np.arange(df[key].shape[0])+1
df[key]['PositiveOutcomeFrequency'] = df[key]['PreviousPositiveOutcomes'] / df[key]['PreviousTotalOutcomes']
df[key]['ChoiceOptimal'] = np.where(df[key]['PositiveOutcomeFrequency'] >= 0.5, 1, 0)
df[key][columns]
```
